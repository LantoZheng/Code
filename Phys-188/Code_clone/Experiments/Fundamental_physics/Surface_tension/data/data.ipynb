```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
data = pd.read_excel('data.xlsx')
```

```python
U = data['U'][:7]
f = data['f'][:7]

# Add a constant column to the independent variable
X = sm.add_constant(f)

# Create the linear regression model
model = sm.OLS(U, X)

# Fit the model to the data
results = model.fit()

# Print the regression results
print(results.summary())

rsquared = results.rsquared
print(f"High-precision R-squared: {rsquared:.8f}")
```

```python
# Plot the fitted curve
plt.plot(f, results.fittedvalues, label='Fitted Curve')

# Plot the original points
plt.scatter(f, U,color='orange', label='Original Points')

# Add labels and title
plt.xlabel('f(N)')
plt.ylabel('U(V)')
plt.title('Fitted U-f Curve with Raw data')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
