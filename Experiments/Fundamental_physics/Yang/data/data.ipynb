```python
import pandas as pd
from scipy.stats import linregress
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_excel('data.xlsx')
```

```python
d = data['d'][0:10]
L = data['L'][0]
H = data['H'][0]
D = data['D'][0]
M = data['M'][0:22]
F = 9.8 * M
x = D * data['x'][0:22] / (2*H)
```

```python
d_mean = d.mean()
print(d_mean)
```

```python
slope, intercept, r_value, p_value, std_err = linregress(x, F)
print(slope, intercept, r_value, p_value, std_err)
```

```python
def lin(x, a, b):
    return a*x + b
linex = np.linspace(x.min(), x.max(), 100)
plt.scatter(x, F,label='Data')
plt.plot(linex, lin(linex, slope, intercept), color='red',label='Linear fit')
plt.xlabel('x [mm]')
plt.ylabel('F[N]')
plt.legend()
plt.show()
print("F/delta_x = ", slope)
```

```python
E = (4*L/(np.pi*(d_mean**2)))*slope*10e6
print(E)
```
