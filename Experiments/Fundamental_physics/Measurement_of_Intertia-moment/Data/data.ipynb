```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
data = pd.read_excel('Data.xlsx')
data_filtered = data.dropna(subset=['T', 'beta'])
import numpy as np
import matplotlib.pyplot as plt

# Extract the 'T' and 'beta' columns from the DataFrame
T = data_filtered['T']
beta = data_filtered['beta']

# Perform linear regression
coefficients = np.polyfit(beta, T, 1)
poly = np.poly1d(coefficients)

# Plot the data and the linear regression line
plt.scatter(beta, T, label='Data')
plt.plot(beta, poly(beta), color='red', label='Linear Regression')
plt.ylabel('T(N·m)')
plt.xlabel('beta(rad/s^2)')
plt.legend()
plt.show()
print("Regression Coefficients:", coefficients)


# Calculate the R-squared value
r_squared = r2_score(T, poly(beta))

print("R-squared:", r_squared)
```
