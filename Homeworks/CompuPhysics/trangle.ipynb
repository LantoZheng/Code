```python
import numpy as np
import matplotlib.pyplot as plt

a = 2/3 * np.pi

def draw(n, l, loc):
    if n == 1:
        str = "F+F+F+"
        A = 0
        for i in range(len(str)):
            if str[i] == 'F':
                new_loc = loc + l * np.exp(1j * A)
                plt.plot([loc.real, new_loc.real], [loc.imag, new_loc.imag])
                loc = new_loc
            elif str[i] == '+':
                A = A + a
    else:
        step = l / 2
        draw(n-1, step, loc)
        draw(n-1, step, loc + step)
        draw(n-1, step, loc + step * np.exp(1j * np.pi/3))

plt.figure()
draw(4, 1, 0)
plt.axis('equal')
plt.show()
```
